# Проект: Обучение с учителем: качество модели

## Цель: 

Удерживать активность постоянных клиентов с помощью персонализированных предложений.


## Задачи:

1. Построить модели, которые предскажут вероятность снижения покупательской активности клиента в следующие три месяца.
2. Выбрать лучшую модель.
3. В исследование нужно включить дополнительные данные финансового департамента о прибыльности клиента: какой доход каждый покупатель приносил компании за последние три месяца.
4. Используя данные модели и данные о прибыльности клиентов, выделить сегменты покупателей и разработать для них персонализированные предложения.

## Описание данных

Для анализа дано 4 датасета:

    1. `market_file.csv` - содержит данные о поведении покупателя на сайте, о коммуникациях с покупателем и его продуктовом поведении.
    2. `market_money.csv` - содержит данные о выручке, которую получает магазин с покупателя, то есть сколько покупатель всего потратил за период взаимодействия с сайтом.
    3. `market_time.csv` - содержит данные о времени (в минутах), которое покупатель провёл на сайте в течение периода.
    4. `money.csv` – содержит данные о среднемесячной прибыли покупателя за последние 3 месяца: какую прибыль получает магазин от продаж каждому покупателю.

###  1. `market_file.csv`

**Описание данных:**

- id — номер покупателя в корпоративной базе данных.
- Покупательская активность — рассчитанный класс покупательской активности (целевой признак): «снизилась» или «прежний уровень».
- Тип сервиса — уровень сервиса, например «премиум» и «стандарт».
- Разрешить сообщать — информация о том, можно ли присылать покупателю дополнительные предложения о товаре. Согласие на это даёт покупатель.
- Маркет_актив_6_мес — среднемесячное значение маркетинговых коммуникаций компании, которое приходилось на покупателя за последние 6 месяцев. Это значение показывает, какое число рассылок, звонков, показов рекламы и прочего приходилось на клиента.
- Маркет_актив_тек_мес — количество маркетинговых коммуникаций в текущем месяце.
- Длительность — значение, которое показывает, сколько дней прошло с момента регистрации покупателя на сайте.
- Акционные_покупки — среднемесячная доля покупок по акции от общего числа покупок за последние 6 месяцев.
- Популярная_категория — самая популярная категория товаров у покупателя за последние 6 месяцев.
- Средний_просмотр_категорий_за_визит — показывает, сколько в среднем категорий покупатель просмотрел за визит в течение последнего месяца.
- Неоплаченные_продукты_штук_квартал — общее число неоплаченных товаров в корзине за последние 3 месяца.
- Ошибка_сервиса — число сбоев, которые коснулись покупателя во время посещения сайта.
- Страниц_за_визит — среднее количество страниц, которые просмотрел покупатель за один визит на сайт за последние 3 месяца.

###  2. `market_money.csv`

**Описание данных:**

- id — номер покупателя в корпоративной базе данных.
- Период — название периода, во время которого зафиксирована выручка. Например, 'текущий_месяц' или 'предыдущий_месяц'.
- Выручка — сумма выручки за период.

###  3. `market_time.csv`

**Описание данных:**

- id — номер покупателя в корпоративной базе данных.
- Период — название периода, во время которого зафиксировано общее время.
- минут — значение времени, проведённого на сайте, в минутах.

###  4. `money.csv`

**Описание данных:**

- id — номер покупателя в корпоративной базе данных.
- Прибыль — значение прибыли.

```python
def data_processing (x): #создал функцию для предварительной подготовки таблиц перед анализом данных. 
    #название столбцов со срочной буквы
    x.columns = x.columns.str.lower() 
    print('______________________')
    print()
    print('\033[1m' + 'ДУБЛИКАТЫ ДО УДАЛЕНИЯ:' + '\033[0m')
    #проверяю на наличие явных дубликатов.
    print(x.duplicated().sum()) 
    print()
    print('______________________')
    #удаляю явные дубликаты и сбрасываю индексы в датафрейме.
    for elem in x:
        x = x.drop_duplicates().reset_index(drop=True) 
    print('\033[1m'+ 'ПРОПУСКИ:'+ '\033[0m')
    print('______________________')
    #проверка на пропуски
    print(x.isna().sum())
    print('______________________')
    print()
    #повторная проверка на дубликаты после их удаления
    print('\033[1m'+ 'ДУБЛИКАТЫ ПОСЛЕ УДАЛЕНИЯ:'+ '\033[0m')
    print(x.duplicated().sum()) 
    print()
    print('______________________')
    print()
    print('\033[1m'+ 'ИНФОРМАЦИИ О ДАТАФРЕЙМЕ:'+ '\033[0m')
    print('______________________')
    #информация о датасете
    print(x.info()) 
    return x
```

```python
#Функция для построения графиков
def graphs_plot_barh(data, x, y, graphs_title): #функция для построения графиков
    data.value_counts().sort_values(ascending=True).plot.barh(figsize=(8, 6), fontsize=13)
    plt.suptitle(graphs_title, fontsize=18)
    plt.xlabel(x, fontsize=15)
    plt.ylabel(y, fontsize=15)
    plt.show()
```

```python
#Функиця для построения гистограмм
def graphs_hist(data, x, y, graphs_title, bins): #функция для построения графиков
    data.hist(figsize=(8, 6), bins=bins)
    plt.suptitle(graphs_title, fontsize=18)
    plt.xlabel(x, fontsize=15)
    plt.ylabel(y, fontsize=15)
    plt.show()
```

```python
#Создал сводную таблицу по каждому периоду в market_money
market_money_pivot = market_money.pivot_table(index=['id'], columns=['период'])
market_money_pivot.columns = ['выручка_предыдущий_месяц', 'выручка_препредыдущий_месяц', 'выручка_текущий_месяц']
```

```python
#Создал сводную таблицу по каждому периоду в market_time
market_time_pivot = market_time.pivot_table(index=['id'], columns=['период'])
market_time_pivot.columns = ['минут_предыдущий_месяц', 'минут_текущий_месяц']
```

```python
#провел объединение таблиц market_file, market_money, market_time
df_full = market_file.join((market_money_pivot, market_time_pivot),  how='inner')
```

### `Общий вывод по объединению датасетов`.

**Для объединения датасетов сначала проверил не отсутствуют ли каки-либо ID в датасетах.**

**Удостоверился, что все датасеты содержать одинакое количество уникальных ID.**

**Назначил индексами данные столбцп 'id' во всех датасетах.**

# Использование пайплайнов

```python
df_full_test = df_full.copy()
```

```python
#Создал постоянную величину для одинакового формирования рандомных выборок при вызове кода
RANDOM_STATE = 42

X = df_full.drop(['покупательская активность'], axis=1)
y = df_full['покупательская активность']
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size = 0.25, 
    random_state = RANDOM_STATE,
    stratify = y)
```

```python
ord_cols = ['тип сервиса','разрешить сообщать']
num_cols = ['маркет_актив_6_мес','маркет_актив_тек_мес','длительность','акционные_покупки',
            'средний_просмотр_категорий_за_визит','неоплаченные_продукты_штук_квартал','ошибка_сервиса','страниц_за_визит',
            'выручка_предыдущий_месяц','выручка_препредыдущий_месяц', 'выручка_текущий_месяц',
            'минут_предыдущий_месяц', 'минут_текущий_месяц']
ohe_cols = ['популярная_категория']
```

```python
ohe_pipe = Pipeline(
    [
        (
            'simpleImputer_ohe', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ohe', 
            OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)
        )
    ]
) 
```

```python
ord_pipe = Pipeline(
    [
        (
            'simple_imputer_ord_before',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ord',
            OrdinalEncoder(categories=[
                ['премиум', 'стандарт'],
                ['да', 'нет']
            ],
                           handle_unknown='use_encoded_value',
                           unknown_value=np.nan)
        ),
        (
            'simple_imputer_ord_after',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
)
```

```python
# создал экземпляр класса LabelEncoder для кодирования целевого признака
label_encoder = LabelEncoder()

# обучил модель и трансформировал тренировочную выборку 
y_train = label_encoder.fit_transform(y_train)

# трансформировал тестовую выборку
y_test = label_encoder.transform(y_test)
```

```python
print(label_encoder.classes_)
```

**Категория клиентов с 'Прежний уровень' приняла значение 0, а 'Снизилась' – 1.**

```python
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_cols),
     ('ord', ord_pipe, ord_cols),
     ('num', MinMaxScaler(), num_cols)
    ], 
    remainder='passthrough'
)
```

```python
# создаём итоговый пайплайн: подготовка данных и модель
pipe_final = Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
    ]
)
```

```python
param_grid = [
    # словарь для модели DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range (5, 10),
        'models__max_features': range (2, 16),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    
    # словарь для модели KNeighborsClassifier() 
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range (6, 19),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']   
    },

    # словарь для модели LogisticRegression()
    {
        'models': [LogisticRegression(
            random_state=RANDOM_STATE, 
            solver='liblinear', 
            penalty='l1',
        )],
        'models__C': range (1, 10),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    # словарь для модели SVC()
    {
        'models': [SVC(random_state=RANDOM_STATE, probability=True)],
        'models__degree': range(2, 10),
        'models__kernel': ['linear', 'rbf', 'poly'],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    }
]
```

```python
randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=5,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
```

```python
randomized_search.fit(X_train, y_train)

print('Лучшая модель и её параметры:\n\n', randomized_search.best_estimator_)
print ('Метрика лучшей модели на кросс-валидации:', randomized_search.best_score_)
```

```python
y_train_pred = randomized_search.predict_proba(X_train)
print(f'Метрика ROC-AUC на тренировочной выборке: {roc_auc_score(y_train, y_train_pred[:, 1])}')
```

```python
y_test_pred = randomized_search.predict_proba(X_test)
print(f'Метрика ROC-AUC на тестовой выборке: {roc_auc_score(y_test, y_test_pred[:, 1])}')
```

```python
dummy_clf = DummyClassifier(strategy="constant", constant = 1, random_state=RANDOM_STATE)
dummy_clf.fit(X_train, y_train)
y_train_pred_dummy = dummy_clf.predict_proba(X_train)
```

```python
print(f'Метрика ROC-AUC на тренировочной выборке DummyClassifier: {roc_auc_score(y_train, y_train_pred_dummy[:, 1])}')
```

```python
y_test_pred_dummy = dummy_clf.predict_proba(X_test)
print(f'Метрика ROC-AUC на тестовой выборке DummyClassifier: {roc_auc_score(y_test, y_test_pred_dummy[:, 1])}')
```

**Сравнение результатов лучшей модели с результатами константной модели DummyClassifier показали, что модель LogisticRegression(C=6, penalty='l1', random_state=42, solver='liblinear') лучше спарвляется с предсказаниями целевого признака.**

Цель: удерживать активность постоянных клиентов с помощью персонализированных предложений

Задача: разработать решение, которое позволит персонализировать предложения постоянным клиентам, чтобы увеличить их покупательскую активность.

Для анализа были доступны данные о поведении покупателя на сайте, о коммуникациях с покупателем и его продуктовом поведении – `market_file.csv`; данные о выручке, которую получает магазин с покупателя, то есть сколько покупатель всего потратил за период взаимодействия с сайтом – `market_money.csv`; данными о времени (в минутах), которое покупатель провёл на сайте в течение периода – `market_time.csv`; данные о среднемесячной прибыли покупателя за последние 3 месяца: какую прибыль получает магазин от продаж каждому покупателю – `money.csv`.

В датасетах не обнаружил явных дубликатов дубликатов, но требовалось поменять формат данных в столбцах и избавится от неявных дубликатов.

Для поиска лучшей модели разработал пайплайн в котором перебирались модели: `DecisionTreeClassifier`, `KNeighborsClassifier`, `LogisticRegression` и `SVC`. Также использовал разные гиперпараметры для перечисленных моделей. Чтобы определить какая модель лучше использовал `RandomizedSearchCV`, методы `.best_score_` и `.best_estimator_`. 

Лучшей моделью стала LogisticRegression(C=2, penalty='l1', random_state=42, solver='liblinear'). С оптимальным порогом для построения предсказаний модели `0,84`. То есть при прогнозировании активности покупателей, будем использовать этот порог как оптимальный для предсказания. Он позволит уменьшить количество ошибок второго рода. 
`recall` модели = 0,886. То есть модель в 88,6% случаев верно определила, что активность клиента снизится, но в 11,4% случаев она ошибается и предсказывает, что активность клиента не снизится, хотя в действительности она снизалась.**

Известно, что покупательская активность больше всего снижается у клиентов, которые посещают от 1 до 7 страниц.
Предлагается поработать с категорией клиентов, для которых предсказано снижение активности с вероятностью более 84%.

Глобальное предложение для всех покупателей во всех категориях товаров – это работать над сокращением количества посещений страниц и увеличения покупательской активности. Сейчас, чтобы найти подходящий товар, покупатель посещает 7 страниц и может не найти подходящего товара. Наблюдается большее снижение покупательской активности у тех, кто посещает до 7 страниц. Предполагается, что рекомендательная система или поисковая выдача не может удовлетворить покупательский запрос за 7 попыток, только если покупатель посещает более 7 страниц, тогда он находит необходимый товар.

Чтобы удерживать активность постоянных клиентов с помощью персонализированных предложений предлагается поработать над персонализированными предложениями для трех категорий покупателей в записимости от популярного сегмента товаров. В частности предлагается сократить число страниц, которое посещают покупатели в категориях популярных товаров `Домашний текстиль`, `Косметика и акусесуары`, `Товары для детей`, так как это самые большие категории в которых зафиксировали более число сокращения активности. Чтобы увеличить число активных клиентов в этих категориях предлагается переработать систему поиска и/или рекомендаций в них и сократить число посещаемых страниц минимум до 4-х. Предполагается, что чем быстрее клиент будет находить нужный товар, тем вероятнее его активность будет сохраняться на прежнем уровне. Для того,что бы сократить количество посещаемых страниц, необходимо разобраться в поведении этих категорий клиетов на сайте. Возможно, стоит поработать над дополнительной системой фильтров для поиска в этих категориях. Это может позволить давать более релевантные ответы на запрос клиента. 

Если смотреть на отдельные категории товаров, то снижение покупательской активности в категории `Домашний текстиль` и посещают до 5 страниц с вероячтностью более 84% снизится покупательская активность. В категориях `Косметика и акусесуары` и `Товары для детей` покупательская активность вероятнее будет снижатся при посещении до 4-х страниц.

Если посмотреть на категори, где зафиксировали наименьшее снижение покупательской активности – это `Мелкая бытовая техника и электроника` и `Техника для красоты и здоровья`, то, вероятно, там покупатели находят быстрее необходимый товар. В этих категориях вероятность снижения активности уменьшается, начиная с 4-й страницы.

Условно можно выделить терпеливых покупателей и нетерпеливых. Терпеливые, чтобы совершить покупку готовы посетить более 7 страниц, а нетерпеливые не готовы посещать более 7 страниц, чтобы найти необходимый товар. Предполагется, что сокращение визитов до 4-х страниц позволит сократить число тех, чья активность будет снижаться.

Дополнительно можно изучить категории клиентов, через опрос на сайте для тех, кто в этих категориях отложили товар в корзину, но не купили его. Спросить по какой причине не совершили покупку.
